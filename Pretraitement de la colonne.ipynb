```python
# on va traiter la colonne address2 du fichier subscriber comme exemple
e = etl.fromcsv('PLES_CI_DT_SUBSCRIBER', delimiter = '|', quotechar = "'", quoting = csv.QUOTE_ALL , encoding="utf8")
```

```python
l=e['address_2']
```

```python
vp = pd.DataFrame(l,columns = ['address_2'])
vp[['Ville','Province']] = vp.address_2.str.split(",",expand=True,)
vp
#On divise la colonne address_2 pour qu'on puisse traiter ville et province independamment
```

```python
ps = PorterStemmer()
for w in vp['Ville']:
    print(w, " : ", ps.stem(w))
```
